```python
import requests
import openai
import os
from time import time,sleep
from getpass import getpass
```

```python
def open_file(filepath):
    with open(filepath, 'r', encoding='utf-8') as infile:
        return infile.read()

def save_file(filepath, content):
    with open(filepath, 'w', encoding='utf-8') as outfile:
        outfile.write(content)
```

```python
elapikey = getpass()
openai.api_key = getpass()
```

```python
%%writefile prompt4.txt
Ignore all previous instructions. ACT as a proffesional secratary. Your task is to write precise and accurate summaries of meetings.

Write a precise and accurate summary from the following {text}:

text = <<STORY>>

WRITE A PRECISE AND ACCURATE SUMMARY FROM THE {TEXT} ABOVE:
```

```python
%%writefile input.txt
What is generative AI?
Despite sounding intimidating, generative AI is a straightforward concept. All it refers to is AI algorithms that generate or create an output, such as text, photo, video, code, data, and 3D renderings, from data they are trained on. The premise of generative AI is to create content, as opposed to other forms of AI, which might be used for other purposes, such as analysing data or helping to control a self-driving car.
ChatGPT has become extremely popular, accumulating more than one million users a week after launching. Seeing this success, many other companies have decided to compete in the generative AI space, including Google, Microsoft's Bing, and Opera. The buzz around generative AI is sure to keep on growing as more companies race to join in and find new use cases.Machine learning refers to the subsection of AI that teaches a system to make a prediction based on data it's trained on. An example of this kind of prediction is when DALL-E is able to create an image based on the prompt you enter by discerning what the prompt actually means. Generative AI is, therefore, a machine-learning framework. 
```

```python
#THIS FUNCTION USES CHATGPT
def gpt_31 (messages):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": messages}]
    )
    text = response['choices'][0]['message']['content']
    return text
            
if __name__ == '__main__':
    main1 = open_file("input.txt")
    storyscene = open_file('prompt4.txt').replace('<<STORY>>', main1)
    completion = gpt_31(storyscene)
    print('\n\n', completion)
    save_file('output.txt', completion)            
    
    #ElvenLabs Voice
    url = "https://api.elevenlabs.io/v1/text-to-speech/21m00Tcm4TlvDq8ikWAM"
    headers = {
        'accept': 'audio/mpeg',
        'xi-api-key': elapikey,
        'Content-Type': 'application/json'
    }
    data = {
        'text': completion,
        'voice_settings': {
            'stability': 0.6,
            'similarity_boost': 0.85
        }
    }
    response = requests.post(url, headers=headers, json=data)
    if response.status_code == 200:
        with open('output.mp3', 'wb') as f:
            f.write(response.content)
        print('Text-to-speech conversion successful')
    else:
        print('Error:', response.text)
```
