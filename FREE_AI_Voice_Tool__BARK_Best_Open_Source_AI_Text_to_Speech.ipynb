* Bark can generate highly realistic, multilingual speech as well as other audio - including music, background noise and simple sound effects. 
* The model can also produce nonverbal communications like laughing, sighing and crying.
* It can do voice cloning

##Basics

```python
from bark import SAMPLE_RATE, generate_audio, preload_models
from IPython.display import Audio
#
preload_models()
```

```python
text_prompt = """
     Hello, my name is Plaban. And, uh — and I am a Data Scientish. [huffff] 
     But I also have other interests such as drawing window shopping.
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

##Advanced Examples

```python
text_prompt = """
    Buenos días Miguel. Tu colega piensa que tu alemán es extremadamente malo. 
    But I suppose your english isn't terrible.
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    WOMAN: I would like an oatmilk latte please.
    MAN: Wow, that's expensive!
"""
audio_array = generate_audio(text_prompt)
Audio(audio_array, rate=SAMPLE_RATE)
```

```python
text_prompt = """
    I have a silky smooth voice, and today I will tell you 
    about the exercise regimen of the common sloth.
"""
audio_array = generate_audio(text_prompt, history_prompt="en_speaker_1")
Audio(audio_array, rate=SAMPLE_RATE)
```
